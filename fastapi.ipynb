```python
heights = np.random.uniform(1.5, 2.0, len(df))
```

```python
weights = []
for i in range(len(df)):
    if not pd.isna(df.iloc[i]['bmi']):
        # Use existing BMI to calculate weight
        weight = df.iloc[i]['bmi'] * (heights[i] ** 2)
        weights.append(round(weight, 1))
    else:
        # If BMI is NaN, generate a reasonable weight
        weights.append(round(np.random.uniform(50, 100), 1))
```

```python
df['height_m'] = np.round(heights, 2)
df['weight_kg'] = weights
```

```python
df['calculated_bmi'] = np.round(df['weight_kg'] / (df['height_m'] ** 2), 1)
```

```python
from sklearn.utils import resample

def oversample_data(df, target_col='stroke'):
    # Separate majority and minority classes
    df_majority = df[df[target_col] == 0]
    df_minority = df[df[target_col] == 1]

    # Upsample minority class to match majority class
    df_minority_upsampled = resample(df_minority,
                                   replace=True,     # sample with replacement
                                   n_samples=len(df_majority),  # match majority class size
                                   random_state=42)

    # Combine majority class with upsampled minority class
    df_balanced = pd.concat([df_majority, df_minority_upsampled])
    return df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_oversampled = oversample_data(df)
print(f"\n2. Oversampling:")
print(f"   New size: {len(df_oversampled)} rows")
print(f"   Distribution: {dict(df_oversampled['stroke'].value_counts())}")
```
